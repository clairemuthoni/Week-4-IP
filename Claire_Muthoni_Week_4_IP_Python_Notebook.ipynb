```
# correcting typos in order to get cleaner data 
df_a['Public_name'] = df_a['Public_name'].str.replace('Paris/EugÃ¨ne OudinÃ©/51', 'Eguisheim/Oudincourt')
```

```
#correcting typos \
df_a['Address'] = df_a['Address'].str.replace('EugÃ¨ne OudinÃ©', 'Eguisheim/Oudincourt')
```

```
df_a['Public_name'] = df_a['Public_name'].str.replace('MÃ©nelotte/8','Macnelotte')
```

```
df_a['Address'] = df_a['Address'].str.replace('MÃ©nelotte','Macnelotte')
```

```
# converting the date into the right format 
df_a['Date'] = df_a[df_a.columns[20:23]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
df_a['Date'] = pd.to_datetime(df_a['Date'])
```

```
df_a['Time'] = df_a[df_a.columns[23:25]].apply(
    lambda x: ':'.join(x.dropna().astype(str)),
    axis=1
)
df_a['Time'] = pd.to_datetime(df_a['Time'])
```
